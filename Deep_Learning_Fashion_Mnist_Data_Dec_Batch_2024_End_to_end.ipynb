**Step-6: Develop ANN**

- Input layer

- Hidden layer

- Output layer

- All are in sequantial

- We are passing image , image is 2D matrix form

- Means we are passing pixel values 784 pixels are passing so flatten 2D to 1D

- Every layes aftet input consider as dense layer

- But starting layer always input , last layer always output and middile layers are hidden

```python
from tensorflow.keras import Sequential
from tensorflow.keras.initializers import GlorotUniform
from tensorflow.keras.layers import Flatten, Dense,BatchNormalization,Dropout
```

```python
model = Sequential()
# input layer   starts flatten 28x28 === 1-D
model.add(Flatten(input_shape = (28, 28)))   # 28x28=784 input
# one Hidden layer
# how many neurons wants to use : 128
# which activation function wants to use : 'relu'
# bcz it avoids vanish gradients problem
model.add(Dense(128, activation = 'relu'))  #
# Output layer
# how many classes are there =10
# how many neuorns : 10
# activation: softmax
model.add(Dense(10, activation = 'softmax'))
```

```python
model = Sequential()
model.add(Flatten(input_shape = (28, 28)))
model.add(BatchNormalization())
model.add(Dropout(0.3))
###########################################################
model.add(Dense(128,
                activation = 'relu',
                kernel_initializer=GlorotUniform()))
model.add(BatchNormalization())
model.add(Dropout(0.3))
##########################################################
model.add(Dense(10,
                activation = 'softmax',
                kernel_initializer=GlorotUniform()))

# Q1) How to apply batch Normaliztion
# Q2) How to apply Drop out
# Q3) Xaviour weights bias intialization
```

**Non Trainable parameter**: There are the parameter are fixed will not change during the training

**Trainable parameters**: These are the parametes are vary change while doing the training

**Step-8: Compile**

```python
from re import VERBOSE
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

**Step-9: Divide train data into train and validation**

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X_train,y_train,test_size=0.2)
```

```python
# 48000 divided into some batches
# Each batch has 64 observations
# Total 48000/64= 750Batches
```

```python
history=model.fit(X_train,
                  y_train,
                  epochs=10,
                  batch_size=64,
                  validation_data=(X_val,y_val))
```
